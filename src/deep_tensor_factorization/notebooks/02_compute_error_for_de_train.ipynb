```python
import os
import pandas as pd
import numpy as np
```

```python
DE_TRAIN = "/data/pinello/PROJECTS/2023_08_ZL/kaggle_scp/data/de_train.parquet"
```

```python
df = pd.read_parquet(DE_TRAIN)
```

```python
df.head()
```

```python
df = df.drop(['sm_lincs_id', 'SMILES', 'control'], axis=1)
df = pd.melt(df, id_vars=['cell_type', 'sm_name'], var_name='gene', value_name='p_value')
df.sort_values(['cell_type', 'sm_name'], inplace=True)
```

```python
df.head()
```

```python
df['predict'] = 0
```

```python
df['diff'] = (df['p_value'] - df['predict']) ** 2
```

```python
df.head()
```

```python
df = df.drop(['gene', 'p_value', 'predict'], axis=1)
```

```python
df.head()
```

```python
grouped = df.groupby(['cell_type', 'sm_name'])
```

```python
df= grouped.mean()
```

```python
np.mean(np.sqrt(df['diff']))
```
