```python
import os
import pandas as pd

import numpy as np
import scanpy as sc
import mudata as md

import warnings
warnings.filterwarnings('ignore')
```

```python
mdata = md.read_h5mu('../../results/preprocessing_data/multiome_train.h5mu')
```

```python
adata = mdata['rna'].copy()
```

```python
adata.layers['counts'] = adata.X.copy()
```

```python
sc.pp.calculate_qc_metrics(adata, inplace=True)
sc.pp.normalize_total(adata)
sc.pp.log1p(adata)
```

```python
sc.pp.highly_variable_genes(adata, batch_key='donor_id')
```

```python
adata = adata[:, adata.var_names[adata.var['highly_variable_nbatches'] > 2]]
```

```python
sc.tl.pca(adata)
sc.pp.neighbors(adata)
sc.tl.umap(adata)
```

```python
sc.pl.umap(adata, color=['donor_id', 'cell_type'])
```
