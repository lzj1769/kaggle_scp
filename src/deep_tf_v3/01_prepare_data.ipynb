```python
import os
import pandas as pd
```

```python
RESULTS_DIR = "/data/pinello/PROJECTS/2023_08_ZL/kaggle_scp/results/deep_tf"
DE_TRAIN = "/data/pinello/PROJECTS/2023_08_ZL/kaggle_scp/data/de_train.parquet"
```

```python
if not os.path.exists(RESULTS_DIR):
    os.makedirs(RESULTS_DIR)
```

```python
df = pd.read_parquet(DE_TRAIN)

df.head()
```

```python
# convert to long df
df = df.sort_values(['cell_type', 'sm_name'])
df = df.drop(['sm_lincs_id', 'SMILES', 'control'], axis=1)
df = pd.melt(df, id_vars=['cell_type', 'sm_name'], var_name='gene', value_name='target')
```

```python
df.head()
```

```python
df.to_csv(os.path.join(RESULTS_DIR, 'train.csv'))
```
